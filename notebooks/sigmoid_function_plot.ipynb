# Sigmoid Function Plot

In this notebook, we will plot the sigmoid function. The sigmoid function is commonly used in machine learning and neural networks as an activation function.

## Import necessary libraries

First, we import the libraries needed for numerical computations and plotting.

```python
# Import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

## Define the sigmoid function

The sigmoid function is defined as \( \sigma(x) = \frac{1}{1 + e^{-x}} \). Let's implement this function in Python.

```python
# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

## Generate data for the plot

We need to generate a range of x values and compute the corresponding sigmoid values.

```python
# Generate x values and compute y values
x = np.linspace(-10, 10, 100)
y = sigmoid(x)
```

## Plot the sigmoid function

Now, we'll plot the sigmoid function using matplotlib.

```python
# Plot the sigmoid function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sigmoid Function')
plt.title('Sigmoid Function Plot')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.legend()
plt.grid(True)
plt.savefig('sigmoid_function_plot.png')
plt.show()
```
